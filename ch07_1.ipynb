```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gimg=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gimg, cmap='gray')
#cv2_imshow(gimg)
```

```python
outimg = np.zeros(gimg.shape, dtype=np.uint8)
cv2_imshow(outimg)
```

```python
scaling = 1.5
for y in range(gimg.shape[0]):
  for x in range(gimg.shape[1]):
    h=int(y/scaling)
    w=int(x/scaling)
    if h < gimg.shape[0] and w < gimg.shape[1]:
      outimg[y,x] = gimg[h,w]
cv2_imshow(outimg)
```

```python
#outimg = cv2.resize(gimg, None, fx=3, fy=3, interpolation=cv2.INTER_NEAREST)
#outimg = cv2.resize(gimg, None, fx=3, fy=3, interpolation=cv2.INTER_LINEAR)
outimg = cv2.resize(gimg, None, fx=3, fy=3, interpolation=cv2.INTER_CUBIC)

cv2_imshow(outimg)
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gimg=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gimg, cmap='gray')
#cv2_imshow(gimg)
```

```python
height, width = gimg.shape
print(height, width)
```

```python
M = cv2.getRotationMatrix2D((gimg.shape[1]/2, gimg.shape[0]/2), 45, 1 )
print(M)
```

```python
outimage=cv2.warpAffine(gimg, M, (width, height))
cv2_imshow(outimage)
```

```python
M= np.float32([[1,0,30], [0,1,20]])
outimage=cv2.warpAffine(gimg, M, (width, height))
cv2_imshow(outimage)
```
