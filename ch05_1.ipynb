```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np
import cv2
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/lena256.jpg')
print(img.shape)
gimg1=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#plt.imshow(gimg1, cmap='gray')
cv2_imshow(gimg1)
```

```python
def saturation(value):
  if value>255:
    return 255
  elif value<0:
    value= 0
  else:
    return value
```

```python
kernnel3 = np.ones((3,3), np.float32)/9
```

```python
output = cv2.filter2D(gimg1, -1, kernnel3)
cv2_imshow(output)
```

```python
gOut = np.zeros((gimg1.shape[0], gimg1.shape[1]), dtype=np.ubyte)
```

```python
for h in range(1,gimg1.shape[0]-1):
  for w in range(1,gimg1.shape[1]-1):
    sum = 0
    for i in range(-1,2):
       for j in range(-1,2):
        sum += gimg1[h+i,w+j]*kernnel3[i+1,j+1]
        sum=saturation(sum)
    gOut[h,w] = sum
cv2_imshow(gOut)
```
