```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/milkdrop.bmp')
print(img.shape)
gimg2=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gimg2, cmap='gray')
```

```python
gPlus=np.zeros((256,256),dtype=np.uint16)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    #gPlus[h,w] = gimg1[h,w] + gimg2[h,w]
    gvalue=np.int32(gimg1[h,w]) + np.int32(gimg2[h,w])
    if gvalue>255:
      gPlus[h,w]=255
    else:
      gPlus[h,w]=gvalue

plt.imshow(np.clip(gPlus, 0, 255).astype(np.uint8), cmap='gray')
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    gPlus[h,w] = gimg1[h,w] * 0.5 + gimg2[h,w] * 0.5


plt.imshow(np.clip(gPlus, 0, 255).astype(np.uint8), cmap='gray')
```

```python
gPlus = cv2.add(gimg1,gimg2)
plt.imshow(np.clip(gPlus, 0, 255).astype(np.uint8), cmap='gray')
```

```python
gMinu = np.zeros((gimg1.shape[0], gimg1.shape[1]), dtype=np.int16)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    gMinu[h,w] = np.abs(np.int32(gimg1[h,w] - np.int32(gimg2[h,w])))


plt.imshow(np.clip(gMinu, 0, 255).astype(np.uint8), cmap='gray')
```
